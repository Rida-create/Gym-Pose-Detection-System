```python
# =====================================================================
# CELL 2: Import Libraries
# =====================================================================
"""
Import all required libraries for pose estimation and visualization.
"""

import cv2
import numpy as np
import mediapipe as mp
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
from google.colab import files
import math
from typing import Tuple, List, Dict, Optional

print("‚úì All libraries imported successfully")
```

```python
# =====================================================================
# CELL 3: Initialize Models
# =====================================================================
"""
Load YOLOv10 for person detection and MediaPipe Pose for landmark extraction.
"""

# Initialize YOLOv10 (using YOLOv8 as placeholder - YOLOv10 uses same API)
print("Loading YOLOv10 model...")
yolo_model = YOLO('yolov8n.pt')  # Nano model for speed
print("‚úì YOLOv10 loaded")

# Initialize MediaPipe Pose
print("Loading MediaPipe Pose...")
mp_pose = mp.solutions.pose
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles

pose_detector = mp_pose.Pose(
    static_image_mode=False,
    model_complexity=1,
    smooth_landmarks=True,
    enable_segmentation=False,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)
print("‚úì MediaPipe Pose loaded")
```

```python
# =====================================================================
# CELL 4: Utility Functions - Angle Calculation
# =====================================================================
"""
Functions for calculating joint angles using vector mathematics.
"""

def calculate_angle(point1: Tuple[float, float],
                   point2: Tuple[float, float],
                   point3: Tuple[float, float]) -> float:
    """
    Calculate angle between three points using vector mathematics.

    Args:
        point1: First point (x, y) - start of first vector
        point2: Second point (x, y) - vertex of angle
        point3: Third point (x, y) - end of second vector

    Returns:
        Angle in degrees (0-180)

    Example:
        For elbow angle: shoulder -> elbow -> wrist
        angle = calculate_angle(shoulder, elbow, wrist)
    """
    # Extract coordinates
    x1, y1 = point1
    x2, y2 = point2
    x3, y3 = point3

    # Calculate vectors
    vector1 = np.array([x1 - x2, y1 - y2])
    vector2 = np.array([x3 - x2, y3 - y2])

    # Calculate angle using dot product
    cosine_angle = np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))

    # Clamp to avoid numerical errors
    cosine_angle = np.clip(cosine_angle, -1.0, 1.0)

    # Convert to degrees
    angle = np.degrees(np.arccos(cosine_angle))

    return angle


def get_landmark_coords(landmarks, landmark_id: int,
                       frame_width: int, frame_height: int) -> Tuple[int, int]:
    """
    Convert normalized MediaPipe landmarks to pixel coordinates.

    Args:
        landmarks: MediaPipe pose landmarks
        landmark_id: ID of the landmark (e.g., mp_pose.PoseLandmark.LEFT_SHOULDER)
        frame_width: Width of the original frame
        frame_height: Height of the original frame

    Returns:
        Tuple of (x, y) pixel coordinates
    """
    landmark = landmarks.landmark[landmark_id]
    x = int(landmark.x * frame_width)
    y = int(landmark.y * frame_height)
    return (x, y)
```

```python
# =====================================================================
# CELL 5: Rep Counter Class
# =====================================================================
"""
Class for counting exercise repetitions and analyzing form.
"""

class ExerciseCounter:
    """
    Tracks exercise repetitions and provides form feedback.
    Supports: squats, bicep curls, and extensible to other exercises.
    """

    def __init__(self, exercise_type: str = "squat"):
        """
        Initialize the exercise counter.

        Args:
            exercise_type: Type of exercise ("squat" or "bicep_curl")
        """
        self.exercise_type = exercise_type
        self.counter = 0
        self.stage = None  # "up" or "down"
        self.feedback = []

        # Exercise-specific thresholds
        if exercise_type == "squat":
            self.down_threshold = 90  # Hip angle threshold for squat down position
            self.up_threshold = 160   # Hip angle threshold for standing
        elif exercise_type == "bicep_curl":
            self.down_threshold = 160  # Elbow angle threshold for extended arm
            self.up_threshold = 40     # Elbow angle threshold for curled position


    def count_squat_rep(self, hip_angle: float, knee_angle: float) -> Dict:
        """
        Count squat repetitions based on hip and knee angles.

        Args:
            hip_angle: Angle at hip joint (degrees)
            knee_angle: Angle at knee joint (degrees)

        Returns:
            Dictionary with counter, stage, and feedback
        """
        self.feedback = []

        # Check if in down position (squat)
        if hip_angle < self.down_threshold and knee_angle < 100:
            if self.stage != "down":
                self.stage = "down"
                # Form feedback for squat depth
                if knee_angle < 70:
                    self.feedback.append("Great depth!")
                elif knee_angle > 90:
                    self.feedback.append("Go deeper")

        # Check if in up position (standing)
        elif hip_angle > self.up_threshold and knee_angle > 160:
            if self.stage == "down":
                self.stage = "up"
                self.counter += 1
                self.feedback.append("Rep counted!")

        return {
            "counter": self.counter,
            "stage": self.stage,
            "feedback": self.feedback
        }


    def count_bicep_curl_rep(self, elbow_angle: float) -> Dict:
        """
        Count bicep curl repetitions based on elbow angle.

        Args:
            elbow_angle: Angle at elbow joint (degrees)

        Returns:
            Dictionary with counter, stage, and feedback
        """
        self.feedback = []

        # Check if arm is extended (down position)
        if elbow_angle > self.down_threshold:
            if self.stage != "down":
                self.stage = "down"
                self.feedback.append("Arm extended")

        # Check if arm is curled (up position)
        elif elbow_angle < self.up_threshold:
            if self.stage == "down":
                self.stage = "up"
                self.counter += 1
                self.feedback.append("Rep counted!")

            # Form feedback
            if elbow_angle < 30:
                self.feedback.append("Full contraction!")

        return {
            "counter": self.counter,
            "stage": self.stage,
            "feedback": self.feedback
        }


    def reset(self):
        """Reset the counter and stage."""
        self.counter = 0
        self.stage = None
        self.feedback = []
```

```python
# =====================================================================
# CELL 6: Visualization Functions
# =====================================================================
"""
Functions for drawing skeleton, angles, and feedback on frames.
"""

def draw_skeleton(frame: np.ndarray, landmarks,
                 mp_pose, mp_drawing) -> np.ndarray:
    """
    Draw pose skeleton on the frame.

    Args:
        frame: Input frame (BGR)
        landmarks: MediaPipe pose landmarks
        mp_pose: MediaPipe pose object
        mp_drawing: MediaPipe drawing utilities

    Returns:
        Frame with skeleton drawn
    """
    mp_drawing.draw_landmarks(
        frame,
        landmarks,
        mp_pose.POSE_CONNECTIONS,
        landmark_drawing_spec=mp_drawing_styles.get_default_pose_landmarks_style()
    )
    return frame


def draw_angle(frame: np.ndarray, point1: Tuple[int, int],
               point2: Tuple[int, int], point3: Tuple[int, int],
               angle: float) -> np.ndarray:
    """
    Draw angle value and arc on the frame.

    Args:
        frame: Input frame
        point1, point2, point3: Three points forming the angle
        angle: Calculated angle value

    Returns:
        Frame with angle annotation
    """
    # Draw angle text near the vertex
    text = f"{int(angle)}"
    cv2.putText(frame, text, (point2[0] - 30, point2[1] - 20),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 0), 2)

    # Draw vertex point
    cv2.circle(frame, point2, 8, (0, 255, 255), -1)
    cv2.circle(frame, point2, 10, (255, 0, 0), 2)

    return frame


def draw_info_panel(frame: np.ndarray, counter: int, stage: str,
                   feedback: List[str], exercise: str) -> np.ndarray:
    """
    Draw information panel with counter, stage, and feedback.

    Args:
        frame: Input frame
        counter: Rep counter value
        stage: Current stage ("up" or "down")
        feedback: List of feedback messages
        exercise: Exercise type

    Returns:
        Frame with info panel
    """
    h, w = frame.shape[:2]

    # Draw semi-transparent panel
    overlay = frame.copy()
    cv2.rectangle(overlay, (10, 10), (400, 200), (0, 0, 0), -1)
    frame = cv2.addWeighted(overlay, 0.6, frame, 0.4, 0)

    # Exercise name
    cv2.putText(frame, f"Exercise: {exercise.upper()}", (20, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)

    # Rep counter
    cv2.putText(frame, f"Reps: {counter}", (20, 80),
                cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 0), 3)

    # Stage
    stage_color = (255, 200, 0) if stage == "down" else (0, 255, 255)
    stage_text = stage.upper() if stage else "START"
    cv2.putText(frame, f"Stage: {stage_text}", (20, 120),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, stage_color, 2)

    # Feedback
    y_offset = 160
    for fb in feedback:
        cv2.putText(frame, fb, (20, y_offset),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (100, 255, 100), 2)
        y_offset += 30

    return frame
```

```python
# =====================================================================
# CELL 7: Main Processing Pipeline
# =====================================================================
"""
Core function that integrates YOLOv10, MediaPipe, and rep counting.
"""

def process_frame(frame: np.ndarray, counter: ExerciseCounter) -> Tuple[np.ndarray, Dict]:
    """
    Process a single frame through the complete pipeline.

    Pipeline:
    1. Detect person using YOLOv10
    2. Extract pose landmarks using MediaPipe
    3. Calculate joint angles
    4. Count reps and analyze form
    5. Draw visualizations

    Args:
        frame: Input frame (BGR)
        counter: ExerciseCounter instance

    Returns:
        Tuple of (processed_frame, info_dict)
    """
    h, w = frame.shape[:2]
    output_frame = frame.copy()

    # Step 1: Detect person using YOLOv10
    results = yolo_model(frame, verbose=False)

    person_detected = False
    person_box = None

    # Find person bounding box
    for result in results:
        boxes = result.boxes
        for box in boxes:
            cls = int(box.cls[0])
            conf = float(box.conf[0])

            # Class 0 is 'person' in COCO dataset
            if cls == 0 and conf > 0.5:
                person_detected = True
                xyxy = box.xyxy[0].cpu().numpy()
                person_box = [int(x) for x in xyxy]

                # Draw bounding box
                cv2.rectangle(output_frame,
                            (person_box[0], person_box[1]),
                            (person_box[2], person_box[3]),
                            (0, 255, 0), 2)
                cv2.putText(output_frame, f"Person {conf:.2f}",
                          (person_box[0], person_box[1] - 10),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
                break
        if person_detected:
            break

    info = {"person_detected": person_detected, "angles": {}, "rep_info": {}}

    if not person_detected:
        cv2.putText(output_frame, "No person detected", (50, 50),
                   cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
        return output_frame, info

    # Step 2: Run MediaPipe Pose on the full frame
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    pose_results = pose_detector.process(frame_rgb)

    if not pose_results.pose_landmarks:
        cv2.putText(output_frame, "Pose not detected", (50, 100),
                   cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 165, 255), 2)
        return output_frame, info

    landmarks = pose_results.pose_landmarks

    # Step 3: Draw skeleton
    output_frame = draw_skeleton(output_frame, landmarks, mp_pose, mp_drawing)

    # Step 4: Extract landmark coordinates (mapped to original frame)
    left_shoulder = get_landmark_coords(landmarks, mp_pose.PoseLandmark.LEFT_SHOULDER, w, h)
    left_elbow = get_landmark_coords(landmarks, mp_pose.PoseLandmark.LEFT_ELBOW, w, h)
    left_wrist = get_landmark_coords(landmarks, mp_pose.PoseLandmark.LEFT_WRIST, w, h)
    left_hip = get_landmark_coords(landmarks, mp_pose.PoseLandmark.LEFT_HIP, w, h)
    left_knee = get_landmark_coords(landmarks, mp_pose.PoseLandmark.LEFT_KNEE, w, h)
    left_ankle = get_landmark_coords(landmarks, mp_pose.PoseLandmark.LEFT_ANKLE, w, h)

    # Step 5: Calculate joint angles
    elbow_angle = calculate_angle(left_shoulder, left_elbow, left_wrist)
    hip_angle = calculate_angle(left_shoulder, left_hip, left_knee)
    knee_angle = calculate_angle(left_hip, left_knee, left_ankle)

    info["angles"] = {
        "elbow": elbow_angle,
        "hip": hip_angle,
        "knee": knee_angle
    }

    # Step 6: Count reps based on exercise type
    if counter.exercise_type == "squat":
        rep_info = counter.count_squat_rep(hip_angle, knee_angle)
        # Draw relevant angles for squats
        output_frame = draw_angle(output_frame, left_shoulder, left_hip, left_knee, hip_angle)
        output_frame = draw_angle(output_frame, left_hip, left_knee, left_ankle, knee_angle)

    elif counter.exercise_type == "bicep_curl":
        rep_info = counter.count_bicep_curl_rep(elbow_angle)
        # Draw elbow angle for bicep curls
        output_frame = draw_angle(output_frame, left_shoulder, left_elbow, left_wrist, elbow_angle)

    info["rep_info"] = rep_info

    # Step 7: Draw info panel
    output_frame = draw_info_panel(
        output_frame,
        rep_info["counter"],
        rep_info["stage"],
        rep_info["feedback"],
        counter.exercise_type
    )

    return output_frame, info
```

```python
# =====================================================================
# CELL 8: Video Processing Function
# =====================================================================
"""
Process video file or webcam stream.
"""

def process_video(video_path: str, exercise_type: str = "squat",
                 output_path: str = "output.mp4", max_frames: int = None):
    """
    Process video file with pose estimation and rep counting.

    Args:
        video_path: Path to input video (or 0 for webcam)
        exercise_type: Type of exercise ("squat" or "bicep_curl")
        output_path: Path to save output video
        max_frames: Maximum number of frames to process (None = all)
    """
    # Initialize counter
    counter = ExerciseCounter(exercise_type)

    # Open video
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print(f"‚ùå Error: Could not open video: {video_path}")
        return

    # Get video properties
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    print(f"üìπ Video Info: {frame_width}x{frame_height} @ {fps} FPS")
    print(f"Total frames: {total_frames}")

    # Initialize video writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

    frame_count = 0

    print("üöÄ Processing video...")

    while True:
        ret, frame = cap.read()

        if not ret:
            break

        frame_count += 1

        # Process frame
        processed_frame, info = process_frame(frame, counter)

        # Write frame
        out.write(processed_frame)

        # Progress update
        if frame_count % 30 == 0:
            print(f"Processed {frame_count}/{total_frames if max_frames is None else min(max_frames, total_frames)} frames | Reps: {counter.counter}")

        # Check max frames limit
        if max_frames and frame_count >= max_frames:
            break

    # Cleanup
    cap.release()
    out.release()

    print(f"‚úÖ Processing complete!")
    print(f"Total reps counted: {counter.counter}")
    print(f"Output saved to: {output_path}")
```

```python
# =====================================================================
# CELL 11: Use Webcam (Advanced - Requires Colab Browser Access)
# =====================================================================
"""
Real-time processing using webcam.
Note: Webcam access in Colab requires special setup using JavaScript.
This is an alternative approach for recorded videos.
"""

from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.8):
    """Capture photo from webcam."""
    js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
    display(js)
    data = eval_js('takePhoto({})'.format(quality))
    binary = b64decode(data.split(',')[1])
    with open(filename, 'wb') as f:
        f.write(binary)
    return filename

# Example usage:
photo_path = take_photo()
frame = cv2.imread(photo_path)
counter = ExerciseCounter("squat")
processed_frame, info = process_frame(frame, counter)
cv2_imshow(processed_frame)

print("üì∏ Webcam capture function ready!")
print("Run: photo_path = take_photo() to capture a photo")
```
